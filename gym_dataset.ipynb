```python
def clean_gym_data(gym_dataset) -> pd.DataFrame:
    
    cols_to_drop = ["Max BPM", "Avg BPM", "Resting BPM"]
    gym_dataset= gym_dataset.drop(columns=cols_to_drop)
    gym_dataset = gym_dataset.dropna()
    return gym_dataset
 
# While this might seem redundant because I dropped a column above, its imperative to show there are 
# multiple ways to drop columns and these in particular also had no value to my datasat
 
```

```python
gym_dataset_cleaned = clean_gym_data(gym_dataset)
gym_dataset_cleaned.head()
#displaying work from code above to show its been cleaned
```

```python
print(gym_dataset_cleaned.mean(numeric_only=True))
print(gym_dataset_cleaned.median(numeric_only=True))
print(gym_dataset_cleaned.mode(numeric_only=True))

#showing you can get mean, median, and mode for the entire dataset
```

```python
def summarize_gymset(gym_dataset_cleaned: "Calories Burned"):
   
   
    gymset_sum = gym_dataset_cleaned['Calories Burned'].sum()
    gymset_mean = gym_dataset_cleaned['Calories Burned'].mean().round(2)
    gymset_median = gym_dataset_cleaned['Calories Burned'].median()

    # Return a formatted string
    return (f"The total number of calories is {gymset_sum}.\n"
            f"The mean number of calories is {gymset_mean}.\n"
            f"The median number of calories is {gymset_median}.")
    
#using this formatting to make the dataset more specific to narrow down the data to specific categories and turning it into a string.
```

```python
summary = summarize_gymset(gym_dataset_cleaned)
print(summary)
```

```python
def summarize_gymset(gym_dataset_cleaned: "Age"):
   
   
    gymset_sum = gym_dataset_cleaned['Age'].sum()
    gymset_mean = gym_dataset_cleaned['Age'].mean().round(2)
    gymset_median = gym_dataset_cleaned['Age'].median()

    # Return a formatted string
    return (f"The total number of clients age is {gymset_sum}.\n"
            f"The mean number of clients age is {gymset_mean}.\n"
            f"The median number of clients age is {gymset_median}.")
    
#repeated a 2nd time but with new information to see the data from different columns
```

```python
summary_of_age = summarize_gymset(gym_dataset_cleaned)
print(summary_of_age)
```

```python
gym_dataset_cleaned["Height/Weight Ratio"]=gym_dataset_cleaned["Height (m)"]/gym_dataset_cleaned["Weight (kg)"]
#started to build this out to use for a plot
```

```python
age_group = gym_dataset_cleaned.groupby("Age")["Workout Frequency (days/week)"].mean()
weight_group = gym_dataset_cleaned.groupby('Gender')["Weight (kg)"].mean()
workout_group = gym_dataset_cleaned.groupby("Workout Type")["Height/Weight Ratio"].mean()
```

```python
age_group
```

```python
age_group.plot(legend=True,title="Gym Dataset",xlabel="Age",ylabel="Workouts per week",color="green")
#used plot line to demonstrate on age and workout per week varied by group. Also wanted to showcase use of color and titles for graphs
```

```python
weight_group.plot.pie(legend=True,title="Gender")
#used pie chart to showcase the weight of each gender and who was "larger"
```

```python
workout_group.plot.bar(legend=True,title="Gym Dataset",xlabel="Workout Group",ylabel="Height/Weight Ratio",color="purple",ylim=(.024, .026))

#used bar graph to show height/weight comparison next to what kind of workouts they do. Used ylim to zoom into the graph more effectively.
#without ylim the bar was almost negligable in differences because the numbers weren't too far off from a birds eye view
```
